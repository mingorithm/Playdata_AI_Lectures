## index이름, column 이름을 기준으로 정렬

- <b style='font-size:1.2em'>sort_index(axis, ascending=True)</b>
    - axis
        - index 이름 기준 정렬(행) : 'index' 또는 0 (기본값)
        - column 이름 기준 정렬(열) : 'columns' 또는 1
    - ascending
        - 정렬방식
        - True(기본): 오름차순, False: 내림차순
    - inplace
        - 원본에 적용 여부
        - False(기본): 변경한 복사본 반환 
        - True : 원본을 변경

|**sum()**|합계|

- **공통 매개변수**
    - skipna=True(기본값)
        - 결측치(NA)를 포함해서 계산할지 여부 설정.
        - True가 기본값으로 결측치(NA)를 제외하고 처리한다. 
        - 결측치 제외하지 않으려면 skipna=False로 설정하는데 결측치가 있으면 결과는 결측치가 된다.
    - axis
        - DataFrame에 적용할 때 계산방향을 설정
            - 0 또는 'index' (기본값): 컬럼 방향으로 집계
            - 1 또는 'columns': 행 방향으로 집계

## aggregate(func, axis=0, \*args, \*\*kwargs) 또는 agg(func, axis=0, \*args, \*\*kwargs)
- DataFrame, Series의 메소드로 **집계결과를 다양한 형태로 묶어서 볼 때** 사용한다.
    - **사용자 정의 집계메소드를 사용할 때도 편리하다.**
- **매개변수**
    - func 
        - 집계 함수 지정
            - 함수명/함수리스트 (문자열)
                - 판다스 제공 집계메소드들은 문자열로, 사용자정의 집계함수는 함수 객체로 전달
            - Dictionary : {'집계할컬럼' : 집계함수, ... }
                - 컬럼별로 다른 집계를 할 경우
    - axis
        - DataFrame에 적용할 때 0 방향으로 계산할 지 1 방향으로 계산할지 설정
            - 0 또는 'index' (기본값): 컬럼 방향으로 집계
            - 1 또는 'columns': 행 방향으로 집계
    - \*args, \*\*kwargs 
        - 함수에 전달할 매개변수. 
        - 집계함수는 첫번째 매개변수로 Series를 받는다. 그 이외의 매개변수가 있는 경우. 

```python
flights['ARR_DELAY'].min(), flights['ARR_DELAY'].max()
```

```python
flights['ARR_DELAY'].aggregate(["min", "max"])
```

```python
flights['ARR_DELAY'].agg(["min", "max"])
```

```python
flights['ARR_DELAY'].mean(skipna=False)
flights['ARR_DELAY'].agg("mean", skipna=False)  # mean 메소드에 skipna 파라미터 값을 전달.
flights['ARR_DELAY'].agg(["mean", "sum"],  skipna=False)
```

```python
# 컬럼별로 다른 집계
# key: 컬럼명, value: 집계함수(들)
d = {
    "DEP_DELAY":"sum",
    "ARR_DELAY":["min", "max"]
}
flights.agg(d)
```

```python
# DEP_DELAY, ARR_DELAY 의 평균, 표준편차
flights[["DEP_DELAY", "ARR_DELAY"]].agg(["mean", "std"])#, axis=1)
```

```python
flights.select_dtypes(exclude="object").agg(["count", "mean", "std",
                                             "min", "median", "max"])
```

```python
# 사용자 정의 집계 함수
### 첫번째 파라미터: 통계량을 구할 Series
### 반환값: 계산한 통계량

# 최대값과 최소값의 차이
def max_min_diff(columns:pd.Series):
    max_v = columns.max()
    min_v = columns.min()
    return max_v - min_v
```

```python
max_min_diff(flights['DEP_DELAY'])
```

```python
flights['DEP_DELAY'].agg(["min", "max", max_min_diff])
```

```python
flights[['DEP_DELAY', 'ARR_DELAY']].agg(["min", "max", max_min_diff])
```

```python
# IQR 값 계산하는 사용자정의 집계함수를 만들어서 
# 'DEP_DELAY', 'ARR_DELAY' 컬럼에 IRQ값을 조회
# IQR: InterQuartileRange  # 3분위 - 1분위
```

```python
def IQR(columns:pd.Series):
    f, t = columns.quantile(q=[0.25, 0.75])
    iqr = t - f # 3분위 - 1분위
    return iqr
```

```python
IQR(flights['DEP_DELAY']) #, IQR(flights['ARR_DELAY'])
```

```python
flights['DEP_DELAY'].agg(IQR)
```

```python
flights[["DEP_DELAY", "ARR_DELAY"]].agg(IQR)
```

# Groupby
- 특정 열을 기준으로 같은 값을 가지는 행끼리 묶어서 group화 한다.
- **~~ 별 집계를 할 때** 사용한다. (성별, 직급별, 등급별 ...)
    - Group으로 묶을 기준 열은 범주형타입(category)을 사용한다.
- **구문**
    - `DF.groupby('그룹으로묶을기준컬럼')`
        - DataFrameGroupby 객체를 반환한다.
        - DataFrameGroupby 객체는 어떤 행끼리 묶여있는지 정보를 가진다. 이 객체의 집계함수를 사용하면 그룹별 집계를 할 수 있다.
            - `DataFrameGroupby객체['집계할 컬럼'].집계함수()`
        - groupby에 여러개의 컬럼을 기준으로 나눌 경우 리스트에 묶어서 전달한다.
        - 집계할 컬럼이 여러개인 경우 리스트로 묶어준다.
    - 집계함수
        - 기술통계 함수들
        - agg()/aggregate()
            - 여러 다른 집계함수 호출시(여러 집계를 같이 볼경우)
            - 사용자정의 집계함수 호출시
            - 컬럼별로 다른 집계함수들을 호출할 경우

```python
flights.AIRLINE.unique()
```

```python
# groupby() 메소드를 이용해서 나누기(group별로 묶어주기).
r = flights.groupby("AIRLINE") #AIRLINE의 값이 같은 행끼리 같은 그룹으로 묶어준다.
print(type(r))
r.groups #어떻게 묶였는지를 확인
# 'AA': [3, 6, 8, 15, 26, 32, 33, 36, 37, 41, 46, 47, 48, 55, 62, 66, 73, 75, 76, 87, 89, 97, 98, 107, 120, 122, 128, 131, 133, 146, 160, 171, 174, 178, 179, 188, 200, 201, 206, 221, 267, 270, 279, 305, 309, 311, 321, 330, 332, 342, 343, 
```

```python
# AIRLINE별 각 컬럼들의 평균값
flights.groupby('AIRLINE').mean(numeric_only=True)
```

```python
# AIRLINE 별 도착지연시간의 평균
flights.groupby('AIRLINE')["ARR_DELAY"].mean()
```

```python
# AIRLINE 별 출발/도착 지연시간의 평균
flights.groupby("AIRLINE")[['DEP_DELAY', 'ARR_DELAY']].mean().sort_values('ARR_DELAY')
```
